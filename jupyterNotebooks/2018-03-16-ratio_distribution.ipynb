```python
n = 100000
sigma1, sigma2 = 0.1, 0.1

minx2, maxx2 = 0.1, 10

x2_0 = linspace(minx2, maxx2, num=n)
ratio = 1.0 - exp(-x2_0/10)
#ratio = 0 * ratio + 0.2

x1 = x2_0 * ratio + randn(n) * sigma1
x2 = x2_0 + randn(n) * sigma2

z = x1 / x2

figure(figsize=(10,5))

plot(x2, z, ls='', marker='.', markersize=1, color='gray')
ax = gca()

for thresh, col in zip([0.1, 0.4], ['#1f77b4', 'orange']):
    idx = where(x1 > thresh)
    plot(x2[idx], z[idx], ls='', marker='.', markersize=1, color=col)

ax.grid()
ax.set_xscale('log')
ax.set_xlim((0.1, max(x2)))
ax.set_ylim((-0.1, 1))
ax.plot(x2_0, ratio)

######

bins = logspace(-1, 1, num=10)
fig = figure(figsize=(10,10))

indices = digitize(x2, bins)
for i in range(0,10):
    ax = subplot(10, 2, 2*i+1)
    ax.set_xlim((-2, 2))
    h_ = ax.hist(z[where(indices == i)], color='gray',
                 normed=True, histtype='step', bins=20)
    ax.text(0.1, 0.7, '{_:.2f}'.format(_=bins[i]), transform=ax.transAxes)

idx = where(x1 > 0.1)
x2 = x2[idx]
z = z[idx]

indices = digitize(x2, bins)
for i in range(0,10):
    ax = subplot(10, 2, 2*i+2)
    ax.set_xlim((-2, 2))
    h_ = ax.hist(z[where(indices == i)], color='#1f77b4',
                 normed=True, histtype='step', bins=20)
```
