```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
n = 500

r = np.abs(np.random.normal(loc=1, size=n)) # r
N = np.abs(np.random.normal(loc=1, size=n)) # N
v = r**0.37 + 0.1 * np.random.normal(loc=0, size=n) # dv

reg = linear_model.LinearRegression()
fr = reg.fit(np.hstack((np.log(r).reshape((n,1)),
                        np.log(N).reshape((n,1)))),
             np.log(v))

print(fr.coef_, fr.intercept_)

fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(3,1,1)
ax.plot(N, v, ls='none', marker='.')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim((1e-2, 4))

ax = fig.add_subplot(3,1,2)
ax.plot(r, v, ls='none', marker='.')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim((1e-2, 4))

ax = fig.add_subplot(3,1,3)
ax.plot(r*N, v, ls='none', marker='.')
ax.set_xscale('log')
ax.set_yscale('log')
_ = ax.set_xlim((1e-2, 4))
```
